# The objective of this worksheet is to test if Spark is working!

```python
import pyspark
```

```python
from pyspark import SparkContext
```

## Create SparkContext

```python
sc = SparkContext("local","Word Count Exmaple")
```

```python
sc
```

## Create an RDD

```python
rdd = sc.textFile("Test.txt")
```

## Check RDD

```python
rdd
```

```python
words = rdd.flatMap(lambda line: line.split(" "))
```

```python
words
```

```python
words.collect()
```

```python
map1 = words.map(lambda w: (w, 1))
```

```python
wordCounts = map1.reduceByKey(lambda x, y: x + y)
```

```python
wordCounts
```

```python
wordCounts.collect()
```

## Experiment with the other transformations and actions

```python
result = wordCounts.filter(lambda x: x[1] > 5).collect()
result
```

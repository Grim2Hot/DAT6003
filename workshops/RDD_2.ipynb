# The objective of this worksheet is to demonstrate how to create RDDs using parallelize method of SparkContext.

```python
import pyspark
```

```python
from pyspark import SparkContext
```

```python
from time import time
```

# Create SparkContext

```python
sc = SparkContext("local[2]","RDDs using Parallelize")
```

```python
data = [10, 20, 25, 30, 35, 20, 45, 20, 65, 70, 90, 85, 75]
```

```python
t0 = time()
rdd = sc.parallelize(data)
duration = time() - t0
```

```python
print('Total time elapsed to create the RDD: ',duration)
```

```python
rdd
```

```python
t0 = time()
output = rdd.reduce(lambda x, y: x + y)
duration = time() - t0
```

```python
print('Total time elapsed to perform reduce: ',duration)
```

```python
output
```

```python
import functools
t0 = time()
output = functools.reduce(lambda x, y: x + y, data)
duration = time() - t0
```

```python
output
```

```python
print('Total time elapsed to perform reduce: ',duration)
```

```python
data1 = ["Cardiff Met University", "Bristol University", "Aberdeen University", "Essex University"]
```

```python
rdd2 = sc.parallelize(data1)
```

```python
t0 = time()
rdd3 = rdd2.map(lambda x: x.split(" "))
duration = time() - t0
```

```python
print('Total time elapsed to perform map operation: ',duration)
```

```python
rdd3.collect()
```

```python
t0 = time()
rdd4 = rdd2.flatMap(lambda x: x.split(" "))
duration = time() - t0
```

```python
print('Total time elapsed to perform flatMap operation: ',duration)
```

```python
rdd4.collect()
```
